#Representing using graph

```python
# Groups the DataFrame 'sp' by "Region," calculates the sum of "Sales" for each group, and plots a bar chart.
sp.groupby("Region")["Sales"].sum().plot.bar()
```

```python
# Groups the DataFrame 'sp' by "Region," calculates the sum of "Profit" for each group, and then it plots a bar chart with blue bars.
sp.groupby("Region")["Profit"].sum().plot.bar(color='red')
```

```python
# Groups the DataFrame 'sp' by "Region," calculates the sum of "Sales" for each group,and then it plots a pie chart with percentage labels.
sp.groupby("Region")["Sales"].sum().plot.pie(autopct="%1.0f%%")
```

```python
# Groups the DataFrame 'sp' by "Region,"it also calculates the sum of "Profit" for each of teh group, and plots pie chart using percentage labels.
sp.groupby("Region")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```
